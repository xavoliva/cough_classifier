```python
X_no = pd.concat([X_no, y_no["Label"]], axis=1)
X_no["type"] = X_no.apply(
    lambda row: "wet" if row["Label"] == 1.0 else "dry", axis=1)

display(X_no)
```

```python
sns.countplot(x='Expert', data=X_no, hue="type")

VIS_PATH = "../../../vis"
plt.savefig(f'{VIS_PATH}/expert_distribution.pdf',
            format='pdf', bbox_inches='tight')
```

```python
# box plots
print(X.columns)
for top, features in FEATURES.items():
    print(top, features)
    if "SNR" in features and not "SNR" in X.columns:
        features.remove("SNR")
    X[features].plot(kind="box", figsize=(15, 15), logy="sym")
```

```python
N = len(X.columns)

fig, axs = plt.subplots(10, 7, figsize=(50, 40))

for i, feature in enumerate(X.columns):
    subplot = axs[i // 7, i % 7]
    X[feature].plot(kind="hist", bins=200, logy="sym", ax=subplot)
```
